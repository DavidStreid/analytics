```python
%matplotlib inline

import matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel('Questionnaire Dataset (Per Game Stats as of Jan 29 2016).xlsx')
dates = np.array([2015, 2014, 2013, 2012, 2011, 2010, 2009, 2008, 2007, 2006, 2005])
season = data['Season'].unique()
```

# 3P-SHOOTERS ANALYSIS
## 3P-shooters are playing fewer minutes
## 3P-shooter percentage is decreasing slightly on average
## 3P-shooter average assists is decreasing slightly on average

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

data['3p_%'] = data['3PM']/data['3PA']  # Adding three-point percentage to data column

average_3p_playing_time = []
average_playing_time = []

avg_3pp = []
avg_ast = []

for s in season: # Fill average_3p_playing_time, average_playing_time, avg_3pp, avg_ast
    s_data = data.loc[data['Season'] == s] 
    avg_3p = s_data['3PA'].mean()
    shooters = s_data.loc[s_data['3PA']>avg_3p]
    
    avg_3pp.append(shooters['3p_%'].mean())
    avg_ast.append(shooters['AST'].mean())
    
    avg_min = s_data['MIN'].mean()
    average_playing_time.append(avg_min)
    
    average_3p_playing_time.append(shooters['MIN'].mean())

# Adding Graphs
m, b = np.polyfit(dates, average_3p_playing_time, 1)
plt.scatter(dates, average_3p_playing_time)
plt.plot(dates, m*dates + b, '-')
plt.xlabel('Season')
plt.ylabel('Average_3p_playing_time')
plt.title('Average playing time of Top 3-point shooters versus Season')
plt.show()

m, b = m, b = np.polyfit(dates, average_playing_time, 1)
plt.scatter(dates, average_playing_time)
plt.plot(dates, m*dates + b, '-')
plt.yticks(np.arange(int(min(average_playing_time))-1, int(max(average_playing_time))+2, 0.5))
plt.xlabel('Season')
plt.ylabel('Average_playing_time')
plt.title('Average playing time versus Season')
plt.show()

m, b = np.polyfit(dates, avg_3pp, 1)
plt.scatter(dates, avg_3pp)
plt.plot(dates, m*dates + b, '-')
plt.xlabel('Season')
plt.ylabel('Average_3point_percentage')
plt.title('Average_3point_percentage versus Season')
plt.show()

m, b = np.polyfit(dates, avg_ast, 1)
plt.scatter(dates, avg_ast)
plt.plot(dates, m*dates + b, '-')
plt.xlabel('Season')
plt.ylabel('Average_assists')
plt.title('Average_assists_for_3p_shooters versus Season')
plt.show()
```

# Breakdown of Increase in 3-point shooting by position

```python
data['3p_%'] = data['3PM']/data['3PA'] # Adding three-point percentage

centers = []  # Track number of players by positions who are shooting more than the average number of 3-point shots
forwards = []
guards = []

for s in season:
    s_data = data.loc[data['Season'] == s]
    avg_3p = s_data['3PA'].mean()
    shooters = s_data.loc[s_data['3PA']>avg_3p]
    
    positions = shooters.groupby('Position').count()['Season'].tolist()
    try:
        positions[-3]       # If there were seasons where no centers took 3-pointers
    except IndexError:
        positions = [0] + positions
        
    centers.append(positions[0])
    forwards.append(positions[1])
    guards.append(positions[2])

m, b = np.polyfit(dates, centers, 1)
plt.scatter(dates, centers)
plt.plot(dates, m*dates + b, '-')
plt.xlabel('Season')
plt.ylabel('3p_centers')
plt.title('Number of 3p_shooting_centers versus Season')
plt.show()
        
m, b = np.polyfit(dates, forwards, 1)
plt.scatter(dates, forwards)
plt.plot(dates, m*dates + b, '-')
plt.xlabel('Season')
plt.ylabel('3p_forwards')
plt.title('Number of 3p_shooting_forwards versus Season')
plt.show()
    
m, b = np.polyfit(dates, guards, 1)
plt.scatter(dates, guards)
plt.plot(dates, m*dates + b, '-')
plt.xlabel('Season')
plt.ylabel('3p_guards')
plt.title('Number of 3p_shooting_guards versus Season')
plt.show()
```

#SmallBall Phenomenon

```python
convert_height = lambda x: int(x[0]) * 12 + (int(x[3]) if len(x)==6 else int(x[3:5]))
data['Height_in'] = data['Height'].apply(convert_height)  # Converting format of height to 

all_height = []
most_played_height = []
superstar_height = []

for s in season:
    s_data = data.loc[data['Season'] == s]
    avg_min = s_data['MIN'].mean()
    std_min = s_data['MIN'].std()

    all_height.append(s_data['Height_in'].mean())
    
    most_played = s_data.loc[s_data['MIN']>avg_min]
    most_played_height.append(most_played['Height_in'].mean())
    
    superstars = s_data.loc[s_data['MIN']>avg_min+std_min]      # Select players who play 1 std above the mean or more minutes
    superstar_height.append(superstars['Height_in'].mean())

    # TODO - Remove outliers (E.g. Centers)
    # TODO - Examine individual teams
    
m, b = np.polyfit(dates, most_played_height, 1)
plt.scatter(dates, most_played_height)
plt.plot(dates, m*dates + b, '-')
plt.xlabel('Season')
plt.ylabel('average_height')
plt.title('average_height versus Season')
plt.show()

m, b = np.polyfit(dates, most_played_height, 1)
plt.scatter(dates, most_played_height)
plt.plot(dates, m*dates + b, '-')
plt.xlabel('Season')
plt.ylabel('most_played_height')
plt.title('Most_played_height versus Season')
plt.show()

m, b = np.polyfit(dates, superstar_height, 1)
plt.scatter(dates, superstar_height)
plt.plot(dates, m*dates + b, '-')
plt.xlabel('Season')
plt.ylabel('superstar_height')
plt.title('superstar_height versus Season')
plt.show()
```
